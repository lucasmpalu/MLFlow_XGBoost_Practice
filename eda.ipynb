```python
import pandas as pd
```

```python
features_base = [
    "radius", "texture", "perimeter", "area", "smoothness",
    "compactness", "concavity", "concave_points", "symmetry", "fractal_dimension"
]

features_name = ["id", "target"]
features_name += [f"{feature}_mean" for feature in features_base] # mean son las medias de cada feature
features_name += [f"{feature}_se" for feature in features_base] # se son los errores estándar de cada feature
features_name += [f"{feature}_worst" for feature in features_base] # worst son los peores valores de cada feature

df = pd.read_csv("./data/wdbc.data", names=features_name)

print(df.head(20))
```

```python
def contar_outliers(df):
    cols_num = df.select_dtypes(include=['float64', 'int64']).columns
    for col in cols_num:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        bajo = (df[col] < (Q1 - 1.5 * IQR)).sum()
        alto = (df[col] > (Q3 + 1.5 * IQR)).sum()
        if bajo > 0 or alto > 0:
            print(f"{col}: {bajo} outliers bajos / {alto} outliers altos")
contar_outliers(df)
```

```python
df.to_csv("./data/wdbc_clean.csv", index=False)
```
